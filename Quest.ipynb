```python
# EMPEZAR ACA
import pandas as pd
# read file
df = pd.read_excel('GSAF5.xls')

# rename columns
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()

# remove last two columns (NaN in all)
df = df.drop(['unnamed:_21','unnamed:_22'],axis=1)

# country, activity, Type
#hip_vars_df = df[['country','activity','type']]
def clean_char(el,char='?'):
    if el.endswith(char):
        return el[:-1]
    else:
        return el

df['country'] = df['country'].fillna('').str.lower().str.strip() # fill na with '', put lowercase and remove spaces
df['country'] = df['country'].apply(clean_char) # YO limpiar, to lower, remove spaces and remove ?
df['country'] = df['country'].str.replace(' (uae)','').str.replace('&','and').str.replace(' (sri lanka)','').str.replace('columbia','colombia').str.replace('coast of africa','africa')

sorted(df['country'].unique())
```

```python
# José 11:45pm

""" 
I added this, so tommorow I can continue on this part

"""

df.dropna(subset=["activity"])

df["activity"] = df["activity"].str.lower()

df['activity'] = df['activity'].replace({'fishing for red fish': 'fishing', 'Fishing boat swamped in a storm':'fishing','free diving (but on surface)': 'free diving'})

```

```python
# José 11:45pm

""" 
I remove the nan variable,replace some variables like ? to Unconfirmed,
and turn all the variables in lower case
"""
# replace nan, ?, Unconfirmed, Invalid with Unverified. Use map.
#df.dropna(subset=["type"], inplace=True)

df["type"] = df["type"].fillna('Unconfirmed')
df["type"] = df["type"].replace({"?": "Unconfirmed", "Unconfirmed": "Unconfirmed", "Unverified": "Unconfirmed", "Under investigation":"Unconfirmed","Questionable":"Unconfirmed"})
df["type"] = df["type"].str.lower()
df["type"].unique()
df["type"].value_counts()
```
